# Capstone_project_SQL Query

```python
import sqlite3
import pandas as pd
```

```python
# Create a SQLite database
conn = sqlite3.connect('movie_database.db')

# Read the CSV files into dataframes
df_A = pd.read_csv('Capstone_Project_Phase_1.csv')
df_B = pd.read_csv('Capstons_project_phase1_part_B.csv')

# Insert data into respective tables
df_A.to_sql('Table_A', conn, if_exists='replace', index=False)
df_B.to_sql('Table_B', conn, if_exists='replace', index=False)

# Create a new table with the desired structure
query_create = """
CREATE TABLE IF NOT EXISTS CombinedTable (
    Sno INTEGER,
    MovieName TEXT,
    DirectorName TEXT,
    Duration TEXT,
    Genre TEXT,
    Rating REAL
);
"""
conn.execute(query_create)

# Insert data from the joined query into the new table
query_insert = """
INSERT INTO CombinedTable (Sno, MovieName, DirectorName, Duration, Genre, Rating)
SELECT
    a.Sno,
    a.Movie_Name AS MovieName,
    a.Directors_Name AS DirectorName,
    a.Durations AS Duration,
    b.Genere AS Genre,
    a.Ratings
FROM Table_A a
JOIN Table_B b ON a.Sno = b.Sno;
"""

conn.execute(query_insert)

# Commit changes and close the connection
conn.commit()
conn.close()
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('movie_database2.db')

# Define the SQL query to find duplicate rows based on specific columns (e.g., MovieName)
query = """
SELECT MovieName, COUNT(*)
FROM CombinedTable
GROUP BY MovieName
HAVING COUNT(*) > 1;
"""

# Execute the query and fetch all the duplicate rows
duplicate_rows = conn.execute(query).fetchall()

# Print the duplicate rows
if len(duplicate_rows) > 0:
    print("Duplicate Data:")
    for row in duplicate_rows:
        print("MovieName:", row[0])
        print("Count:", row[1])
        print("=======================")
else:
    print("No duplicate data found.")

# Close the connection
conn.close()
```
