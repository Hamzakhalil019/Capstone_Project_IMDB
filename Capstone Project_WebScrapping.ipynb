```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
import requests
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1')
time.sleep(3)
```

```python
Sno=[]         #--1 
Movie_name=[]  #--2
duration=[]    #--3
year=[]        #--4
rating=[]      #--5
meta=[]        #--6
directors=[]   #--7

start=0
end=36
for page in range(start,end):
    
    #Sno--1
    try:
        sno_tags=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
        for i in sno_tags:
            Sno.append(i.text.strip()[0:-1].replace('.','').replace(',',''))
    except NoSuchElementException:
        Sno.append(None)
    
    #Movie_name--2
    try:
        movie_tags=driver.find_elements(By.XPATH,'//h3/a')
        for i in movie_tags:
            Movie_name.append(i.text)
    except NoSuchElementException:
        Movie_name.append(None)
        
    #duration--3
    try:
        dur_tags=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
        for i in dur_tags:
            duration.append(i.text)
    except NoSuchElementException:
        duration.append(None)
        
    #year--4
    try:
        year_tags=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
        for i in year_tags:
            year.append(i.text)
    except NoSuchElementException:
        year.append(None)
        
    #rating--5
    try:
        rat_tags=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
        for i in rat_tags:
            rating.append(i.text)
    except NoSuchElementException:
        rating.append(None)
    
    #directors--6
    try:
        dir_tags= driver.find_elements(By.XPATH,".//p[@class='']")
        for i in dir_tags:
            direct=i.text.strip().split(" | ")[0]
            direct=direct.replace("Director: ",'').replace("Directors: ",'')
            directors.append(direct)
    except NoSuchElementException:
        directors.append(None)
    
    #Metascore--7
    score_tags = driver.find_elements(By.XPATH, '//div[@class="ratings-bar"]')
    for i in score_tags:
        met = i.text
        metascore = None

    # Extract Metascore value if available
        if 'Metascore' in met:
            metascore = met.split("\n")[-1].strip().split()[0]

        meta.append(metascore)
            
   
        
    
    #next button--8
    try:    
        nxt_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        nxt_button.click()
        time.sleep(3)
    except NoSuchElementException:
        break
```

```python
print('Sno :', len(Sno))
print('Movie_name :', len(Movie_name))
print('Duration :', len(duration))
print('Year :', len(year))
print('Rating :', len(rating))
print('Directors :', len(directors))
print('Metascore :', len(meta))
```

```python
data=({'Sno':Sno,'Movie_Name':Movie_name,'Directors_Name': directors,'Durations':duration,'Year':year,'Ratings':rating,'MetaScore':meta})
df_1=pd.DataFrame(data)
df_1
```

```python
import selenium
import requests
from selenium import webdriver

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"Chromedriver.exe")
driver.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
time.sleep(5)
```

```python
sno=[]           #--1
Movie_name=[]    #--2
votes=[]         #--3
gross=[]         #--4
certificates=[]  #--5
actors=[]        #--6
gener=[]        #--7

start=0
end=36
for page in range(start,end):
    
    #Sno--1
    try:
        sno_tags=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
        for i in sno_tags:
            sno.append(i.text.strip()[0:-1].replace('.','').replace(",",''))
    except NoSuchElementException:
        sno.append(None)
    
    #Movie_name--2
    try:
        movie_tags=driver.find_elements(By.XPATH,'//h3/a')
        for i in movie_tags:
            Movie_name.append(i.text)
    except NoSuchElementException:
        Movie_name.append(None)
        
    #Votes---3
    try:
        vot_tags=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
        for tag in vot_tags:
            text = tag.text.strip()
            parts = text.split('|')  # Split the text using '|' as separator
    
    # Find the 'Votes' part and remove rest part--
            votes_part = [part.strip() for part in parts if 'Votes' in part]
        
            if votes_part:
                votes_value = votes_part[0].replace('Votes:','').replace(',','')
                votes.append(votes_value)
            else:
                votes.append(None)  # If no 'Gross' value found, append None
    except NoSuchElementException:
        votes.append(None)
    

    #Gross collection---4
    
    try:
        gro_tags = driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]')

        for tag in gro_tags:
            text = tag.text.strip()
            parts = text.split('|')  # Split the text using '|' as separator
    
    # Find the 'Gross' part and remove everything else--
            gross_part = [part.strip() for part in parts if 'Gross' in part]
    
            if gross_part:
                gross_value = gross_part[0].replace('Gross:','').replace(',','')
                gross.append(gross_value)
        
            else:
                gross.append(None)  # If no 'Gross' value found, append None
    
    except NoSuchElementException:
        gross.append(None)
    
    
   # Certificate----5
    wait = WebDriverWait(driver, 3)
   
    # Wait for the certificate elements to become visible
    cer_tags = wait.until(EC.visibility_of_all_elements_located((By.XPATH, '//p[@class="text-muted "]')))

    # Find elements for certificate

    for cer_tag in cer_tags:
        cer_elements = cer_tag.find_elements(By.XPATH, './/span[@class="certificate"]')
        cert = None
        for cer_element in cer_elements:
            cert_text = cer_element.text.strip()
            if cert_text:  # Check if the text is not empty
                cert = cert_text
                break
        certificates.append(cert if cert else None)  # Append 'nan' for missing values

 
    #Stars--6
    try:
        dir_tags= driver.find_elements(By.XPATH,".//p[@class='']")
        for i in dir_tags:
            actors.append(i.text.strip().split(" | ")[-1].replace("Stars: ", ""))
    except NoSuchElementException:
        actors.append(None)
        
    #Genere--7
    try:
        ge_tag=driver.find_elements(By.XPATH,'//span[@class="genre"]')
        for i in ge_tag:
            gener.append(i.text.strip().split("|")[-1])
    except NoSuchElementException:
        gener.append(None)
    
    

    #next button--
    try:    
        nxt_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        nxt_button.click()
        time.sleep(5)
    except NoSuchElementException:
        break
```

```python
print('Sno :', len(sno))
print('Movie_name :', len(Movie_name))
print('Votes :', len(votes))
print('Gross :', len(gross))
print('Certificates :', len(certificates))
print('Stars :', len(actors))
print('Genere :', len(gener))
```

```python
data=({"Sno":sno,"Movie_Name":Movie_name,'Stars':actors,'Votes':votes,'Genere':gener,'Gross_Collection':gross,'Certification':certificates})
df_A=pd.DataFrame(data)
df_A.head(5)
```
