# **Sub-Query**

**1) Display all details from both tables where movie names are the same.**

```python
combined_tables=pd.merge(df_A,df_B, on= ['Sno','Movie_Name'])
combined_tables
```

**2)   Display all movie names, Director, ratings, and gross collection where the genre is action.**

```python
combined_tables.info()
```

```python
genere_filter = combined_tables['Genere'].str.contains('Action', case=False)
genere_list = combined_tables[genere_filter]

for index, row in genere_list.iterrows():
    print('Sno:',row['Sno'])
    print('Movie:', row['Movie_Name'])
    print('Directors:',row['Directors_Name'])
    print('Ratings:',row['Ratings'])
    print('Gross_Collections:', row['Gross_Collection'])
    print('---------------------------------------------')
```

**3)      Display all details from both tables with the highest gross collection.**

```python
# Assuming 'Gross_Collection' contains values in the format '$X.XXM' where X is a number
# Remove the dollar sign and 'M', and then convert to numeric
combined_tables['Gross_Collection'] = combined_tables['Gross_Collection'].str.replace('$', '').str.replace('M', '').apply(pd.to_numeric, errors='coerce')

# Find the maximum gross collection
max_gross_collection = combined_tables['Gross_Collection'].max()
```

```python
highest_gross_movies = combined_tables[combined_tables['Gross_Collection'] == max_gross_collection]

# Display all details with the highest gross collection
highest_gross_movies
```

```python
for index, row in highest_gross_movies.iterrows():
    print('Sno:',row['Sno'])
    print('Movie:', row['Movie_Name'])
    print('Directors:',row['Directors_Name'])
    print('Durations:',row['Durations'])
    print('Year:', row['Year'])
    print('Ratings:',row['Ratings'])
    print('MetaScore:', row['MetaScore'])
    print('Stars_Name:', row['Stars'])
    print('Votes:', row['Votes'])
    print('Genere:',row['Genere'])
    print('Gross_Collections: $',row['Gross_Collection'],'M')
    print('Certification:',row['Certification'])
    print('Popularity:',row['Popularity'])
    print('---------------------------------------------')
```

```python
combined_tables['Gross_Collection'].max()
```

```python
# Assuming 'Gross_Collection' column is in numeric format
# Convert numeric values to the original format ('$X.XXM')
combined_tables['Gross_Collection'] = combined_tables['Gross_Collection'].apply(lambda x: f"${x:.2f}M" if not np.isnan(x) else np.nan)

# Display the 'Gross_Collection' column with the original format
print(combined_tables['Gross_Collection'])
```

**4)      Display all details from both tables with the highest ratings**

```python
x=combined_tables['Ratings'].max()
y=combined_tables[combined_tables['Ratings']==x]
y
for index, row in y.iterrows():
    print('Sno:',row['Sno'])
    print('Movie:', row['Movie_Name'])
    print('Directors:',row['Directors_Name'])
    print('Durations:',row['Durations'])
    print('Year:', row['Year'])
    print('Ratings:',row['Ratings'])
    print('MetaScore:', row['MetaScore'])
    print('Stars_Name:', row['Stars'])
    print('Votes:', row['Votes'])
    print('Genere:',row['Genere'])
    print('Gross_Collections:',row['Gross_Collection'])
    print('Certification:',row['Certification'])
    print('Popularity:',row['Popularity'])
    print('---------------------------------------------')
```

**5)      Display all details from both tables with the lowest gross collection and lowest ratings**

```python
# Assuming 'Gross_Collection' is in the format '$X.XXM' where X is a number
# Convert 'Gross_Collection' to numeric format
combined_tables['Gross_Collection'] = combined_tables['Gross_Collection'].str.replace('[\$,M]', '', regex=True).astype(float)

# Assuming 'Ratings' column contains numeric ratings
# Convert 'Ratings' to numeric format
combined_tables['Ratings'] = pd.to_numeric(combined_tables['Ratings'], errors='coerce')

# Find the minimum gross collection
min_gross_collection = combined_tables['Gross_Collection'].min()

# Find the minimum ratings
min_ratings = combined_tables['Ratings'].min()

# Filter rows with the minimum gross collection and minimum ratings
lowest_gross_lowest_ratings_movies = combined_tables[(combined_tables['Gross_Collection'] == min_gross_collection) & (combined_tables['Ratings'] == min_ratings)]

# Display all details with the lowest gross collection and lowest ratings
print(lowest_gross_lowest_ratings_movies)
```
